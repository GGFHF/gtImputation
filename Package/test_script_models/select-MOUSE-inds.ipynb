```python
values_list = []
for index, value in enumerate(matriz_df.loc[0, :].values.flatten().tolist()):
    values_list.append([index, value])
values_list
```

```python
selectec_indices_list = []
for index, _ in sorted_values_list[:98]:
    selectec_indices_list.append(index)
selectec_indices_list
```
